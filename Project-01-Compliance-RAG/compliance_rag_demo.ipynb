### *Automated Document Auditing & Risk Assessment*

**Business ROI:** This system reduces manual audit time by **90%** by transforming static PDF policies into an interactive, searchable intelligence layer.

**Core Capabilities:**
* **Instant Gap Analysis:** Compare new contracts against company standards.
* **Source-Backing:** Every answer includes page-level citations to prevent hallucinations.
* **Risk Scoring:** Automatically flags non-compliant clauses.

```
# Install the Google-specific LangChain package
!pip install -qU langchain-google-genai

import os
from google.colab import userdata

# Load the Gemini key
os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')
print("âœ“ Gemini Brain is ready!")
```

```
from langchain_google_genai import ChatGoogleGenerativeAI

# Adding 'latest' usually fixes the 404 SDK error
llm = ChatGoogleGenerativeAI(model="gemini-1.5-flash-latest")

def compliance_check(policy, clause):
    prompt = f"POLICY: {policy}\nCLAUSE: {clause}\nTASK: Does this comply? Answer in one sentence."
    # Using the updated invoke method
    response = llm.invoke(prompt)
    return response.content

# Test it
policy_text = "All employees must use Multi-Factor Authentication (MFA)."
contract_text = "The vendor will use a standard password."

print("ğŸ” ANALYZING WITH GEMINI...")
print("-" * 30)
print(compliance_check(policy_text, contract_text))
```
